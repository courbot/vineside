# Notebook 3. Colorimetry

In this notebook, we investigate the colorimetry of images, focusing on the red/blue balance that is induced by the FasGa staining used in our experiment. Ou method first cluster pixels, then extract colors from regions that are not black/gray/white, to produce a relative color index.

```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile
import segmentation_tools as seg

%load_ext autoreload
%autoreload 2
```

First, let us load the image of interest. Here and in the paper we use x4 cross-section images.

```python
image = tifffile.imread('./input images/Sample 11.2-A-Gw-I-Month 1-x4.tif')
plt.imshow(image)
plt.axis('off')
plt.tight_layout()
```

The results are obtained below, as well as the SLIC over-segmentation used as a processing step.

```python
_,rgb_color,im_slic = seg.get_colors(image) # absolute value
avg_color = rgb_color.mean(axis=0)/rgb_color.mean() # relative index
```

```python
plt.imshow(im_slic);plt.axis('off')
plt.title('SLIC over-segmentation / Relative colors : R=%.2f, G=%.2f, B=%.2f'%(avg_color[0],avg_color[1],avg_color[2]))
plt.tight_layout()
```
